Sidebar: most introductions to matplotlib will have you make your first plot with the function plt.plot(). I personally think this is not a great starting point, as plt.plot() does not give you as much flexibility when your plots become more complex. In this tutorial, we will start with plt.subplots(): there will be a slightly steeper learning curve, but I firmly believe that the way subplots() are implemented will be more beneficial to you in the long run. You can find the docs for plt.subplots() [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html).

Let's start off by creating our Figure and Axes objects. Think of a Figure object as the entire canvas. Multiple Axes can be placed on the Figure (meaning you can have multiple plots on a single Figure), but to keep things simple we'll start off with only one. The reason why I am starting you off with learning how to use plt.subplots() is because it is MUCH easier to make complex changes to an Axes object than it is to plt.plot().

```python
fig, ax = plt.subplots()
```

Here, we've stored our Figure object in 'fig', and our Axes object in 'ax'. If we want to see our plot, we need to call "fig" (see below). We use "ax" to plot data and change plotting parameters - more on that later.

When we call fig, we're basically making a blank canvas. I recommend that you always call fig in the same code block as the rest of your plotting code - things get kind of weird if you don't. You'll see below that we're calling fig in the code block again: we're creating a blank canvas. You can think of each 'ax' call as a layer on the canvas: by calling ax, we're painting over that canvas with additional plotting code. 

Let's make your first plot. We can plot data by calling [ax.plot()](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html) (this is a link!). It's always good to read the documentation to understand how you use tools from modules you import. In the case of ax.plot(), one makes a plot by writing the line ax.plot(X, Y), where X are your x values and Y are your y values.

```python
fig, ax = plt.subplots()

x = [5, 7, 8, 9, 10]
y = [11, 12, 13, 14, 15]

ax.plot(x, y)
```

### kwargs

Now, let's build off this plot and explore the other features that matplotlib has to offer. If you look at the documentation for the .plot() routine, you'll notice a term called 'kwargs'. Without getting too far into the nitty gritty, kwargs is short for 'keyword arguments', and are optional cherries on top for your plots (optional is doing a lot here, you will almost never create plots without using any kwargs).

Here's an example of the most common kwargs:

- color: what color you want to plot in. Common colors (like 'red') can be specified with words, otherwise you can use any color's hex code.
- linestyle: if you are plotting a line, what style do you want the line to be (solid, dashed, dot-dashed)? [Linestyle options listed here](https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_linestyle).
    - linewidth: width of the line.
- marker: if you are plotting points, what markers do you want to use for those points (circles, squares, stars)? [Marker options listed here](https://matplotlib.org/stable/api/_as_gen/matplotlib.markers.MarkerStyle.html#matplotlib.markers.MarkerStyle).
    - markersize = size of the marker.
    - markeredgewidth: width of the marker's edge.
    - markeredgecolor: color of the marker's edge.
- alpha: opacity of the line/marker (1.0 = full, 0.0 = completely transparent).
- label: the label you want to assign to what you're plotting. This won't show up unless you create a legend, which can be done with ax.legend().
- zorder: order in which points are displayed, lower numbers will be displayed underneath higher zordered plot calls.

```python
fig, ax = plt.subplots()

x = [5, 7, 8, 9, 10]
y = [11, 12, 13, 14, 15]

ax.plot(x, y,
        color = '#500000',
        linestyle = '--',
        linewidth = 1.0,
        label = 'Line',
        zorder = 2)

ax.plot(x, y,
        color = 'white',
        marker = '*',
        markersize = 10,
        markeredgecolor = 'k',
        markeredgewidth = 1.0,
        label = 'Markers',
        zorder = 1)

ax.legend()
```

### Formatting axes

You now know how to format lines, markers, and legends - what about the axes? In the example below, we'll be focusing on two of the most common modifications you'll need to apply to axes: the limits and the labels. As always, we should plot our points first. Then, we call the 'ax' object to change the axes.

```python
fig, ax = plt.subplots()

x = [5, 7, 8, 9, 10]
y = [11, 12, 13, 14, 15]

# PLOTTING
ax.plot(x, y,
        color = '#500000',
        linestyle = '--',
        linewidth = 1.0,
        label = 'Line',
        zorder = 2)

ax.plot(x, y,
        color = 'white',
        marker = '*',
        markersize = 10,
        markeredgecolor = 'k',
        markeredgewidth = 1.0,
        label = 'Markers',
        zorder = 1)

# AXES PARAMETERS
ax.set_xlim(4.5, 8.5)
ax.set_ylabel('Y label text')
```

Again, each of the routines we just introduced has kwargs - READ THE DOCUMENTATION!!! The docs aren't just helpful for exploring the options a routine like ax.set_xlabel() provides: **they also list different routines you can use and apply to your plots**. Want to change the tick size? Want to plot vertical and horizontal lines, or text? Want to have labels and numbers on all four sides of your plot? All of these are routines within matplotlib, and will be found in the docs!

### Saving figures

A last step you should be aware of is how to save your plots. Yes, you've created the figure in Jupyter, but you'll probably need to send the plot to your boss or put it in your paper. Doing this is very straightforward: just use plt.savefig().

The argument within plt.savefig() should be the directory you want to save your plot to, plus the plot's name. If you do not include a path (and write only a file name), the plot will be saved in the same directory as your Jupyter notebook.

### subplots

Earlier in the tutorial, we initialized our fig object with plt.subplots(). I'm going to briefly show you why I think this is the best way to plot - plt.subplots() gives you a lot of flexibility.

In the examples above, we didn't provide plt.subplots() with any arguments, so we only created one plot. However, if you read the [plt.subplots() docs](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html), you'll notice that you can specify a number of rows, and number of columns. Here's a practical example:

```python
nrows = 1
ncols = 3

fig, (ax1, ax2, ax3) = plt.subplots(nrows, ncols)
```

Going back to our canvas analogy:

- We create a blank canvas with fig.
- This time, however, we initialize 3 different layers we can paint on: ax1, ax2, and ax3.
- ax1, ax2, and ax3 can all be treated indepently - they are their own plots.
- Parameters of each SUBPLOT is controlled through ax1, ax2, or ax3, while parameters of the entire PLOT as a whole is controlled through fig.

As a quick example: I will plot something on each plot individually with ax1, ax2, and ax3, and then change the overall plot size with one of fig's routines.

```python
nrows = 1
ncols = 3

fig, (ax1, ax2, ax3) = plt.subplots(nrows, ncols)

# Here, I plot on each individual subplot.
ax1.text(0.3, 0.8, 'high text', color = 'red')
ax2.text(0.3, 0.5, 'middle text', color = 'green')
ax3.text(0.3, 0.2, 'low text', color = 'blue')

# Here, I change the overall size of the entire plot.
fig.set_size_inches(10, 3)
```
